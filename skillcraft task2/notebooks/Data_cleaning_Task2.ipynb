CHECKING FOR MISSING VALUES

```python
#checking missing values in each coloum
df.isnull().sum()
```

```python
row_with_missing=df.isnull().any(axis=1).sum()
print(row_with_missing)
```

**HANDLING MISSING VALUES**

```python
#handling the missing values in column
df["Postal Code"] = df["Postal Code"].fillna(
    df.groupby(["City", "State"])["Postal Code"].transform(
        lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x)
    )
)
```

```python
df.isnull().sum()
```

```python
#count the dublicates
df.duplicated().sum()
```

**CHECKING DATA TYPES**

```python
#check current data types
df.dtypes
```

**HANDLING DATA TYPES**

```python
#changing the data types r
df["Order Date"]=pd.to_datetime(df["Order Date"],format="%d-%m-%y",errors="coerce")
df["Ship Date"]=pd.to_datetime(df["Ship Date"],format="%d-%m-%y",errors="coerce")
df["Postal Code"]=df["Postal Code"].astype(str)
```
